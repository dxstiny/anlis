```python
import sympy as sp
from anlis.multidimensional.differential import totalDifferential, absoluteDifferential
from IPython.display import display

u, v = sp.symbols("u v")
du, dv = sp.symbols("du dv")
g = u**2 + u*v

x0 = (1, 2)
x1 = (1.2, 2.1)

dg = totalDifferential(g, (du, dv), u, v) # (a)

change = dg.subs({u: x0[0], v: x0[1], du: x1[0] - x0[0], dv: x1[1] - x0[1]}) # (b)
# or
change = absoluteDifferential(g, x0, (x1[0] - x0[0], x1[1] - x0[1]), u, v) # (b)

display(dg, change)
```

```python
import sympy as sp
from anlis.multidimensional.contour import contourTangent

x, y, z = sp.symbols('x y z')

gradient = ( 2 + 0 * x,
            -5 + 0 * y,
            -1 + 0 * z)

x0 = (1, 3, 7)

contourTangent(gradient, x0, (x, y, z))
```

```python
import sympy as sp
from anlis.multidimensional.derivative import newtonRaphson

x, y = sp.symbols('x y')
F = ( x - y + 1, x**2 - y -1 )
x0 = (1, 1)

x, y = newtonRaphson(F, (x, y), x0)

float(x), float(y)
```

```python
import sympy as sp
from anlis.multidimensional.derivative import partialCompositeDerivative

x, y, t = sp.symbols('x y t')

z = x * y ** 2
fx = sp.Eq(x, sp.exp(-t))
fy = sp.Eq(y, sp.sin(t))

partialCompositeDerivative(z, (fx, fy), t)
```

```python
import sympy as sp
from anlis.multidimensional.derivative import partialDerivativeN

x, y = sp.symbols('x y')
f = y * sp.sin(x) - x * sp.sin(y)

partialDerivativeN(f, (x, y, y)), partialDerivativeN(f, (y, x, y)), partialDerivativeN(f, (y, y, x))
```

```python
import sympy as sp
import matplotlib.pyplot as plt
from anlis.multidimensional.criticalPoints import criticalPoints, maxima, minima
from anlis.multidimensional.contour import contourPlot

x, y = sp.symbols('x y')
f = 2*x**2 - 3*x*y + 8*y**2 + x - y


plt.contour(*contourPlot(f))
criticalPoints(f, x, y), maxima(f, x, y), minima(f, x, y)
```
