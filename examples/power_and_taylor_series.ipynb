```python
import sympy as sp
from anlis.series.power import determine
from anlis.convergence import radiusOfConvergence

x, k = sp.symbols("x k")
f = 5**(k+2) / (k + 3) * x**(k+2)

ak = determine(f, x, k)

radiusOfConvergence(ak, k)
```

```python
import sympy as sp
from anlis.series.taylor import TnMin

x = sp.Symbol("x")
f = sp.cos(x)

I = (-1, 1)

TnMin(f, 0, *I, 5e-4)
```

```python
import sympy as sp
from anlis.series.taylor import taylorPlot

x = sp.Symbol("x")
f = sp.cos(x)

taylorPlot(f, 0, 6, (-1, 1))
```
